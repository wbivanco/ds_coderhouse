```python
df_cpp = df_conta['residencia_provincia_nombre'].value_counts()
```

```python
df_cpp = pd.DataFrame(df_cpp)
```

```python
df_cpp.rename(columns={'residencia_provincia_nombre':'cantidad'}, inplace=True)
```

```python
df_cpp['residencia_provincia_nombre'] = df_cpp.index
```

### Provincias

```python
df_prov = pd.read_csv('datasets/provincias.csv')

df_prov
```

```python
df_prov['nam'].unique()
```

```python
df_prov['nam'] = df_prov['nam'].replace('Ciudad Autónoma de Buenos Aires', 'CABA')
df_prov['nam'] = df_prov['nam'].replace('Tierra del Fuego, Antártida e Islas del Atlántico Sur', 'Tierra del Fuego')
```

```python
df_prov.rename(columns={'nam': 'residencia_provincia_nombre'}, inplace=True)
```

```python
df_prov = df_prov.drop(['gid', 'entidad', 'objeto', 'fna', 'gna', 'in1', 'fdc', 'sag'], axis=1)
```

```python
df_prov.info()
```

```python
df_prov.head()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from geopandas import GeoDataFrame
from shapely.geometry import Point
```

```python
#map_data = df_1.drop('geometry', axis=1)
df_prov.rename(columns={'geom':'geometry'}, inplace=True)
#map_data['gid'] = map_data['gid'].astype('float')
df_prov['geometry'] = gpd.GeoSeries.from_wkt(df_prov['geometry'])
```

```python
df_prov.info()
```

```python
df_cpp
```

```python
df_1 = pd.merge(df_cpp, df_prov, on='residencia_provincia_nombre')
```

```python
df_1
```

```python
df_1.dtypes
```

```python
df_1 = df_1[['cantidad', 'geometry']]
```

```python
df_1.shape
```

```python
gdf = GeoDataFrame(df_1)
```

```python
gdf.plot(figsize=(20, 50))
plt.show()
```
