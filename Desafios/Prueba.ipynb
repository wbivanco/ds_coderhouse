```python
df_prov = pd.read_csv('datasets/provincias.csv')

df_prov
```

```python
df_prov['nam'].unique()
```

```python
df_prov['nam'] = df_prov['nam'].replace('Ciudad Autónoma de Buenos Aires', 'CABA')
df_prov['nam'] = df_prov['nam'].replace('Tierra del Fuego, Antártida e Islas del Atlántico Sur', 'Tierra del Fuego')
```

```python
df_prov.rename(columns={'nam': 'residencia_provincia_nombre'}, inplace=True)
```

```python
# Cargar la capa temática
url1 = 'datasets/provincias.csv'
#natalidad = "datos/natalidad.geojson"

map_data = gpd.read_file(url1)

#ap_data['gid'] = map_data'gid'].astype(float)
#puntos = pd.DataFrame(d)
map_data.info()
```

```python
map_data = map_data.drop('geometry', axis=1)
map_data.rename(columns={'geom':'geometry'}, inplace=True)
map_data['gid'] = map_data['gid'].astype('float')
map_data['geometry'] = gpd.GeoSeries.from_wkt(map_data['geometry'])
```

```python
map_data.info()
```

```python
from geopandas import GeoDataFrame
from shapely.geometry import Point

gdf = GeoDataFrame(map_data)

#geo_ptal_lsoas = gpd.GeoDataFrame(ptal_lsoas)
```

```python
gdf.plot(figsize=(6, 6))
plt.show()

#self.ax = gpd.GeoDataFrame(geometry=[]).plot(figsize=(18, 12))

```
