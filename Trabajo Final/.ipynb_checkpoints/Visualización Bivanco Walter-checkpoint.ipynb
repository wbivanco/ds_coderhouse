### Visualización

### Matplotlib

Las gráficas que vamos a aplicar con la librería Matplotlib son:

* Fecha de apertura del caso con Plot
* Nombre de la provincia de residencia con Pie Chart
* Edad con Boxplot

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

###### Fecha de apertura del caso con Plot

```python
# Agrupo por fecha de apertura del caso para poder contar los registros.
df = contagiados_covid.groupby("fecha_apertura").count()
```

```python
fig, ax = plt.subplots()

# Con el dataframe resultante genero la data con los indices y sus valores.
ax.plot(df.index, df.edad.values, 'go--')
```

###### Nombre de la provincia de residencia con Pie Chart

```python
# Agrupo por provincia y cuento.
total_por_provincia = contagiados_covid.groupby("residencia_provincia_nombre").count()
total_por_provincia = total_por_provincia.id_evento_caso
```

```python
fig, ax = plt.subplots(figsize=(14, 9))

# Paso los valores y los índices del dataframe resultante, y agrego los porcentajes con un decimal al piechart.
ax.pie(total_por_provincia.values, labels=total_por_provincia.index, autopct="%0.1f %%")

plt.title('Cantidad de contagios por provincias')
```

###### Edad con Boxplot

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.boxplot(x=contagiados_covid.edad.index, showfliers=True)

ax.set_title('Boxplot de las edades')

ax.set_xlabel('Edad')
```

### Seaborn

Las gráficas que vamos a aplicar con la librería Seaborn son:

* Sexo por edad con ScatterPlot
* Edad con Histoplot
* Fallecidos por provincia con Barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

###### Sexo por edad con Scatterplot

```python
df = contagiados_covid.copy()
```

```python
f, axs = plt.subplots(nrows=1, ncols=1, figsize=(8, 4))

sns.scatterplot(data=df, x="sexo", y="edad")
```

###### Edad con Histplot

```python
sns.histplot(data=df, x="edad")
```

###### Fallecidos por provincia con Barplot

```python
# Filtro por provincia y cuento
total_por_provincia = contagiados_covid.groupby("residencia_provincia_nombre").count()

# Ordeno el dataset de forma descendente y solo selecciono 15 provincias para que se vean bien en el gráfico.
total_por_provincia = total_por_provincia.sort_values('id_evento_caso', ascending=False)[:15]
```

```python
sns.set_theme(style="whitegrid")

f, axs = plt.subplots(nrows=1, ncols=1, figsize=(18, 4))

sns.barplot(x=total_por_provincia.index, y=total_por_provincia.id_evento_caso, data=total_por_provincia)
```
